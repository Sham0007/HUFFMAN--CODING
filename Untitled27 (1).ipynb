```python
import cv2
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))


def enhance_and_detect(img_gray):
# 1) contrast enhance
img_eq = clahe.apply(img_gray)


# 2) structuring elements
se_long = cv2.getStructuringElement(cv2.MORPH_RECT, SE_SIZE_LONG)
se_sq = cv2.getStructuringElement(cv2.MORPH_RECT, SE_SIZE_SQ)


# 3) top-hat and bottom-hat
opening = cv2.morphologyEx(img_eq, cv2.MORPH_OPEN, se_long)
top_hat = cv2.subtract(img_eq, opening)


closing = cv2.morphologyEx(img_eq, cv2.MORPH_CLOSE, se_long)
bottom_hat = cv2.subtract(closing, img_eq)


# 4) morphological gradient (using square se for gradient)
se_grad = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
dil = cv2.dilate(img_eq, se_grad)
ero = cv2.erode(img_eq, se_grad)
grad = cv2.subtract(dil, ero)


# 5) combine cues (weighted sum)
combined = cv2.addWeighted(grad, 0.6, top_hat, 0.2, 0)
combined = cv2.addWeighted(combined, 1.0, bottom_hat, 0.2, 0)


# 6) threshold
_, bw = cv2.threshold(combined, THRESH_GRAD, 255, cv2.THRESH_BINARY)


# 7) closing to connect line fragments (use elongated se)
bw_closed = cv2.morphologyEx(bw, cv2.MORPH_CLOSE, se_long)


# 8) remove small components
lbl = label(bw_closed > 0)
mask = np.zeros_like(bw_closed)
for region in regionprops(lbl):
if region.area >= MIN_AREA:
rr, cc = region.coords[:,0], region.coords[:,1]
mask[rr, cc] = 255


# 9) optional thinning to get centerlines (useful for visualization)
# skeleton = morphology.skeletonize(mask > 0)


return img_eq, grad, top_hat, bottom_hat, combined, mask




def overlay_and_save(orig, mask, out_path):
# overlay red mask on grayscale image
color = cv2.cvtColor(orig, cv2.COLOR_GRAY2BGR)
red = np.zeros_like(color)
red[:,:,2] = mask
overlay = cv2.addWeighted(color, 0.7, red, 0.3, 0)
cv2.imwrite(out_path, overlay)




if __name__ == '__main__':
files = sorted([f for f in os.listdir(INPUT_FOLDER) if f.lower().endswith(('.png','.jpg','.jpeg','.tif'))])
for fname in tqdm(files):
img = cv2.imread(os.path.join(INPUT_FOLDER, fname), cv2.IMREAD_GRAYSCALE)
if img is None:
continue
img_eq, grad, top_hat, bottom_hat, combined, mask = enhance_and_detect(img)
overlay_and_save(img, mask, os.path.join(OUTPUT_FOLDER, 'det_'+fname))


print('Processing complete. Results in', OUTPUT_FOLDER)
```
